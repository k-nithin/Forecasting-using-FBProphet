```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
 
# Read train.csv
train = pd.read_csv('UPS.csv')
 
# Convert the Date column to datetime type
train['Date'] = pd.to_datetime(train['Date'])
```

```python
store1_dept1_sales = train[(train['Store'] == (22)) & (train['Dept'] == (12))][['Date', 'Weekly_Sales']]
store1_dept1_sales.columns = ['ds', 'y']
print(store1_dept1_sales.shape)
len(store1_dept1_sales)
```

```python
 
# Filtering out only store #1 and department #1 to a new dataframe
# with the required column names


filename = 'StoreDeptWeeklySales.csv'
import os
if os.path.exists(filename):
    os.remove(filename)
# Opening the file and append the data
f = open(filename, 'a')

forecast = pd.DataFrame()


for sid in range(45):
    for did in range(99):
        store1_dept1_sales = train[(train['Store'] == (sid+1)) & (train['Dept'] == (did+1))][['Date', 'Weekly_Sales']]
        store1_dept1_sales.columns = ['ds', 'y']
        if len(store1_dept1_sales) > 5:
            model = Prophet()
            model.fit(store1_dept1_sales)
            future = model.make_future_dataframe(periods=4, freq='W')
            forecast = model.predict(future)
            data = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)
            f.write('Store_'+str(sid+1)+' Dept_'+str(did+1)+'\n')
            data.to_csv(f, header=False, mode = 'a')
            f.write('-----------------------------\n')
            
            
```

```python
# Ploting the predicted weekly sales for Store45 dept 90
figure = model.plot(forecast)
plt.show()
```
